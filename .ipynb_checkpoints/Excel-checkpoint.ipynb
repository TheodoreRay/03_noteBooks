### <span style="color:blue"> 1. Fichiers .txt </span> <br>
> #### <i> stocker les données dans une variable </i>

```python
mylines = []
with open ('test.txt', 'rt') as myfile:
    for myline in myfile:
        mylines.append(myline)
```

> #### <i> affichage ligne par ligne </i>

```python
with open ('test.txt', 'rt') as myfile:
    for line in myfile:
        mylines.append(line)
    for element in mylines:
        print(element, end='') #le paramètre end permet d'éviter d'afficher \n (saut de ligne)
```

> #### <i> recherche d'éléments </i>

```python
mylines[0].find("e",2,20) #recherche la première occurence de la lettre "e" à partir du 3ème caractère ("c") jusqu'au 20ème
```

### <span style="color:blue"> 2. Fichiers Excel </span> <br>
> #### <i> convertir un fichier excel en DataFrame </i>

```python
df = pd.read_excel(r'test.xlsx')
# le tableau étant converti en DataFrame, toutes les méthodes associées sont utilisables :
print(df['variable 1'])
```

> #### <i> interface toute faite pour importer un fichier excel </i>

```python
import tkinter as tk
from tkinter import filedialog
import pandas as pd

root= tk.Tk()

def getExcel (root):
    global df
    import_file_path = filedialog.askopenfilename()
    df = pd.read_excel (import_file_path)
    return df # DataFrame contenant les données du tableau

df = getExcel(root)
root.destroy()
root.mainloop()
```

> #### <i> mise en forme du DataFrame créé </i>

```python
#1) re-définition des colonnes : 1ère ligne
columns = []
for L in range(len(df.iloc[0])):
    columns.append(df.iloc[0, L])
df.columns = columns

#2) re-définition des index : colonne mois
df.index = df['Mois']

#3) suppression des lignes (drop) et colonnes (pop) en doublon
df.drop(['Mois'], inplace=True)
df.pop('Année'); df.pop('Mois')

#4) affichage
df
```
