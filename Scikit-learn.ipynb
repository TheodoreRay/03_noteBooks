### <span style="color:blue"> 1. Régression </span>
> #### <i> a. Moindres carrés (Least Squares) </i>

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
X = [[0, 2], [3, 10], [1, 7], [-2, 4]]
y = [2, 4, 7, 9]
reg = linear_model.LinearRegression()

#méthodes
reg.fit(X, y) #(X, Y) tq Y[0]=a*X[0] +b*X[1]
reg.score(X, y) #coefficient de détermination
```

```python
#attributs
reg.coef_ #(a,b)
```

```python
ypred = [X[i][0]*reg.coef_[0]+X[i][1]*reg.coef_[1] for i in range(4)]

plt.plot(y, label ='yth')
plt.plot(ypred, label = 'ypred')
plt.legend()
plt.grid()
```
