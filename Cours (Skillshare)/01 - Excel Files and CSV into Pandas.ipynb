# Excel files and practical case

Loading data into python from standard Excel files and testing relationships for model building.

## How To

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("DOSNON arrets mensuels.xlsx")

#suppression des lignes concernant des mois sans arrêt
df.drop(df[df['NBR_ARRETS_THERMIQUES']==0].index, inplace=True); df.head(5)
```

```python
total = {'NBR_ARRETS_THERMIQUES': df['NBR_ARRETS_THERMIQUES'].sum(), 'PERTES_ENERGIE' : df['PERTES_ENERGIE'].sum(), 'DUREE_ARRET' : df['DUREE_ARRET'].sum(), }
total
```

## Plot and correlation analysis

```python
ax = plt.stem(df['MOIS_ANNEES'], df['PERTES_ENERGIE'])
ax.set_xticks(df['MOIS_ANNEES'])
plt.grid()
plt.ylabel('pertes d énergie (kW)')
```

```python
sns.pairplot(df)
```

```python
sns.heatmap(df.corr(), square=True)
```
